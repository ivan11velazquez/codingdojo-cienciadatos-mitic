
* Para cargar el dataset, se requiere de la libreria io, el cual contiene el metodo StringIO()
* El parametro pasado a la funcion to_datetime() es df['fechas'] cuando deberia ser df['fecha']
* Luego de ambas correcciones, se compara los graficos con los datos comprobandose que van acorde
* Para mejorar la visualizacion, se ajusto el tamanio de las figuras
* Agregado rotacion de 45 grados a las etiquetas del eje x


```python
import pandas as pd
import matplotlib.pyplot as plt
import io

# Datos de ejemplo en formato CSV
data = """
fecha,ventas
2023-01-01,200
2023-01-02,220
2023-01-03,210
2023-01-04,205
2023-01-05,215
2023-01-06,225
"""

# Cargar los datos en un DataFrame de Pandas
#df = pd.read_csv(pd.compat.StringIO(data))
df = pd.read_csv(io.StringIO(data))

# Convertir la columna 'fecha' a formato datetime
df['fecha'] = pd.to_datetime(df['fecha'])  # Corregido: 'fechas' -> 'fecha'


# Grfico de lneas de las ventas
plt.figure(figsize=(10, 6))
plt.plot(df['fecha'], df['ventas'], marker='o')
plt.title("Ventas Diarias - Grfico de Lneas")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Crear un grfico de dispersin
plt.figure(figsize=(10, 6))
plt.scatter(df['fecha'], df['ventas'], color='red', alpha=0.6)
plt.title("Ventas Diarias - Grfico de Dispersin")
plt.xlabel("Fecha")
plt.ylabel("Ventas")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Crear un histograma de las ventas
plt.figure(figsize=(10, 6))
plt.hist(df['ventas'], bins=5, color='skyblue', edgecolor='black')
plt.title("Distribucin de Ventas - Histograma")
plt.xlabel("Ventas")
plt.ylabel("Frecuencia")
plt.tight_layout()
plt.show()
```
