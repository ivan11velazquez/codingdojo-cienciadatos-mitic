```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

def analisis_graficos(df):

    df['Score'].dropna()

    # variable score
    plt.hist(df['Score'], bins=5, color='skyblue', edgecolor='black', alpha=0.7)
    plt.title("Frecuencia de variable Score")
    plt.xlabel("Puntaje de felicidad")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

 
    plt.figure(figsize=(8, 6))  # Opcional: define el tamaño de la figura
    plt.boxplot(df['Score'], notch=False)
    plt.title("Boxplot Básico")
    plt.ylabel("Valores")
    plt.show()

    # variable GDP per capita
    plt.hist(df['GDP per capita'], bins=5, color='orange', edgecolor='black', alpha=0.7)
    plt.title("Frecuencia de variable GDP per Capita")
    plt.xlabel("GDP per capita")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    # variable Healthy life expectacy
    plt.hist(df['Health (Life Expectancy)'], bins=5, color='green', edgecolor='black', alpha=0.7)
    plt.title("Frecuencia de variable Health (Life Expectancy)")
    plt.xlabel("Esperanza de vida")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()

    # variable Freedom to make life choices
    plt.hist(df['Freedom to make life choices'], bins=5, color='red', edgecolor='black', alpha=0.7)
    plt.title("Frecuencia de variable Freedom to make life choices")
    plt.xlabel("libertad de descision")
    plt.ylabel("Frecuencia")
    plt.grid(True)
    plt.show()



def exploracion_inicial(df):
    # Inicializa una lista para almacenar la información de cada columna
    column_info = []

    # Recorre cada columna para extraer información relevante
    for column in df.columns:
        col_data = {
            'Columna': column,
            'Tipo de Dato': df[column].dtype,                 # Tipo de dato de la columna
            'Valores Nulos': df[column].isnull().sum(),       # Conteo de valores nulos
            'Valores Únicos': df[column].nunique(),           # Conteo de valores únicos
            'Ejemplo de Valores': df[column].unique()[:5],    # Ejemplo de primeros 5 valores únicos
            'Promedio': df[column].mean() if pd.api.types.is_numeric_dtype(df[column]) else None,  # Promedio para columnas numéricas
            'Mediana': df[column].median() if pd.api.types.is_numeric_dtype(df[column]) else None, # Mediana para columnas numéricas
            'Moda': df[column].mode()[0] if not df[column].mode().empty else None                 # Moda de la columna
        }
        column_info.append(col_data)

    # Convierte la lista de diccionarios en un DataFrame para visualización mejorada
    df_summary = pd.DataFrame(column_info)

    # Configura Pandas para que muestre todas las columnas sin truncar
    pd.set_option('display.max_columns', None)
    pd.set_option('display.max_rows', None)
    pd.set_option('display.max_colwidth', 50)  # Ajusta el ancho de las columnas si es necesario

    # Muestra el resumen en consola
    print(df_summary)


df = pd.concat( 
    map(pd.read_csv, ['../../data/world_happiness_report/2015.csv', '../../data/world_happiness_report/2016.csv',
    '../../data/world_happiness_report/2017.csv', '../../data/world_happiness_report/2018.csv',
    '../../data/world_happiness_report/2019.csv']), 
    ignore_index=True)

#exploracion_inicial(df)


print(df['Score'].isna().sum())  # Para verificar si hay valores nulos

#analisis_graficos(df)
#analisis_graficos(df)



def plot_distribution_analysis(data, variables, figsize=(15, 4)):
    """
    Crea una visualización combinada de histograma y boxplot para cada variable.
    
    Parámetros:
    data: DataFrame con los datos
    variables: Lista de variables a analizar
    figsize: Tamaño base de la figura para cada variable
    """
    
    # Configuración del estilo
    plt.style.use('seaborn')
    
    for var in variables:
        # Crear figura con dos subplots
        fig, (ax1, ax2) = plt.subplots(2, 1, figsize=figsize, 
                                      gridspec_kw={'height_ratios': [2, 1]})
        fig.suptitle(f'Análisis de Distribución: {var}', fontsize=14, y=1.05)
        
        # Histograma
        sns.histplot(data=data, x=var, ax=ax1, 
                    color='skyblue', 
                    edgecolor='black',
                    alpha=0.7)
        ax1.set_title('Histograma')
        
        # Boxplot
        sns.boxplot(data=data, x=var, ax=ax2,
                   color='lightgreen',
                   width=0.7)
        ax2.set_title('Boxplot')
        
        # Ajustar espaciado
        plt.tight_layout()
        
        # Calcular estadísticas descriptivas
        stats = data[var].describe()
        outliers = data[data[var] > stats['75%'] + 1.5 * (stats['75%'] - stats['25%'])][var].count()
        
        print(f"\nEstadísticas para {var}:")
        print(f"Media: {stats['mean']:.2f}")
        print(f"Mediana: {stats['50%']:.2f}")
        print(f"Desviación estándar: {stats['std']:.2f}")
        print(f"Rango intercuartil: {stats['75%'] - stats['25%']:.2f}")
        print(f"Número de valores atípicos superiores: {outliers}")
        
        plt.show()

# Lista de variables a analizar
variables = ['Score', 'GDP per capita', 'Healthy life expectancy', 'Freedom to make life choices']

plot_distribution_analysis(df, variables)
```
